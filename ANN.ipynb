```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df=pd.read_csv("/content/Navy Bays_pima-indians-diabetes.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(),annot=True)
```

```python
df.columns
```

```python
x=df.drop(columns=['class'],axis=1)
y=df['class']
#print(x)
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
df.columns
```

```python
df.dtypes
```

```python
model=Sequential([Dense(8,activation='relu',input_dim=8),
                  Dense(6,activation='relu'),
                  Dense(1,activation='sigmoid')
                  ])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(x_train, y_train, epochs=50, batch_size=10, verbose=1)
```

```python
print(model.summary())
```

```python
loss,accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: ",accuracy)
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model , to_file= 'abc.png', show_shapes = True , show_layer_names = True )
```
